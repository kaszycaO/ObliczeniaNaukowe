##  Zadanie 1

Napisać funkcję rozwiązującą równanie f(x) = 0 metodą bisekcji.

```julia
function mbisekcji(f, a::Float64, b::Float64, delta::Float64, epsilon::Float64)
    """ Funkcja rozwiązująca równanie f(x) = 0 metodą bisekcji
    
    Dane wejściowe:
        f: funkcja f(x) zadana jako funkcja anonimowa
        a, b: końce przedziału początkowego
        delta, epsilon: dokładność obliczeń 
    
    Dane wyjściowe w postaci: (r,v,it,err)
        r: szukane przybliżenie
        v: wartość funkcji w punkcie r
        it: liczba iteracji
        err: informacja o błędzie:
            0: brak błędu
            1: źle zadany przedział, funkcja nie zmienia znaku w podanym przedziale [a, b]
    
    """     
    
    
    
    u::Float64 = f(a)
    v::Float64 = f(b)
    e::Float64 = b - a
    
    r::Float64 = 0.0
    w::Float64 = 0.0
    err = 0
    it = 0
    
    if signbit(u) == signbit(v)
        err = 1
        return (r, w, it, err)
    end
    
    while abs(e) > delta
        
        it += 1
        e /= 2.0
        r = a + e
        w = f(r)
        
        if abs(e) < delta || abs(w) < epsilon
           return (r, w, it, err)
        end
        
        if signbit(w) != signbit(u)
            b = r
            v = w
        else
            a = r
            u = w
        end
    end
         
end
```

### Opis

Metoda bisekcji - połowienia, stowrzona na podstawie twierdzenia Darboux, opiera się na dwóch założeniach: podana funkcja musi być ciągła oraz musi zmieniać znak na podanym przedziale $[a, b]$. Początek powyższego algorytmu to deklaracja potrzebnych zmiennych oraz sprawdzenie wcześniej wspomnianego warunku: wartości f(a) - $u$ oraz f(b) - $v$ mają różnić się znakami. Jeżeli warunek ten nie jest spełniony, zwracany jest, zgodny ze specyfikacją, kod błędu: 1, kończąc tym samym wykonywanie algorytmu. Następnie wykonywana jest pętla $while$ z odpowiednim dla metody bisekcji warunkiem końca - wielkość przedziału jest mniejsza niż podana $delta$. W każdej iteracji pętli wyznaczany jest środek wcześniejszego przedziału $[a,b]$ (e/=2.0; r = a + e) oraz liczona jest wartość funkcji w "nowym" środku (w = f(r)). Samo wyznaczanie środka przebiega w bezpieczny sposób, tj. unikamy odejmowania, które mogłoby prowadzić do błędów (klasyczny sposób to $\frac{1}{2}(b-a)$). Jeżeli otrzymany przedział jest dostatecznie mały (|e| < $delta$ ) lub osiągnięto zadaną dokładność (|w| < $epsilon$) algorytm kończy działanie zwracając poprawny wynik. W przypadku, gdy te warunki nie są spełnione, z dwóch powstałych podprzedziałów ($[a, r]$ lub $[r, b]$) wybierany jest ten, w którym funkcja zmienia znak i to on będzie rozpatrywany jako nowy przedział w kolejnej iteracji.

## Zadanie 2

Napisać funkcję rozwiązującą równanie f(x) = 0 metodą Newtona
    

```julia
function mstycznych(f, pf, x0::Float64, delta::Float64, epsilon::Float64, maxit::Int)
     """ Funkcja rozwiązująca równanie f(x) = 0 metodą Newtona
    Dane wejściowe:
        f: funkcja f(x) zadana jako funkcja anonimowa
        pf: pochodna funkcji f
        x0: przybliżenie początkowe
        delta, epsilon: dokładność obliczeń 
        maxit: maksymalna liczba iteracji
    
    Dane wyjściowe w postaci: (r,v,it,err)
        r: szukane przybliżenie
        v: wartość funkcji w punkcie r
        it: liczba iteracji
        err: informacja o błędzie:
            0: brak błędu, metoda zbieżna
            1: nie osiągnięto wymaganej dokładności w podanej liczbie iteracji 
            2: pochodna bliska zeru
    
    """ 
    
    v::Float64 = f(x0)
    r::Float64 = 0.0
    
    
    if abs(pf(x0)) < epsilon
        return (x0 ,v, 0, 2) 
    end
    
    for it = 1:maxit
        
       x1::Float64 = x0 -v/pf(x0)
        v = f(x1)
        
        if abs(x1 - x0) < delta || abs(v) < epsilon
            return (x1 ,v, it, 0)
        end
        x0 = x1   
    end
    
    return (r ,v, maxit, 1)
    
end
```

### Opis

Metoda stycznych (Newtona) jest metodą iteracyjną, korzystającą z linearyzacji - przybliżeniu funkcji nieliniowej za pomocą układu liniowego. W przypadku tego zadania, linearyzacja opiera się na rozwinięciu podanej funkcji w szereg Taylora (2 pierwsze składniki), tj. $f(x) \approx f(x_n) + f'(x_n)(x - x_n)$. W ten sposób otrzymamy przybliżenie funkcji w okolicy argumentu $x_n$. 

Na samym początku algorytm sprawdza, czy pochodna podanej funkcji nie jest bliska zeru w podanym $x_0$. Jeżeli tak jest, to metoda ta jest niemożliwa do użycia i algorytm kończy działanie z kodem błędu: 2. W przeciwnym razie wykonywana jest pętla, $maxit$ razy. W każdej iteracji, liczone jest przybliżenie pierwiastka, za pomocą wzoru $x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$ (dla użytych oznaczeń w algorytmie, $x_n$ = $x_0$). Jeżeli otrzymane przybliżenie jest wystarczająco dokładne, algorytm kończy działanie zwracając poprawny wynik. W przeciwnym wypadku w następnej iteracji będzie liczone przybliżenie dla wyznaczonych wartości ($x_0 = x_1$). Możliwy jest także przypadek, gdy liczba iteracji okaże się za mała na wyznaczenie pierwiastka. Wtedy algorytm kończy działanie z ostatnio wyliczinymi wartościami i kodem błędu: 1.

## Zadanie 3

Napisać funkcję rozwiązującą równanie f(x) = 0 metodą siecznych

```julia
function msiecznych(f, x0::Float64, x1::Float64, delta::Float64, epsilon::Float64, maxit::Int)
     """ Funkcja rozwiązująca równanie f(x) = 0 metodą siecznych
    
    Dane wejściowe:
        f: funkcja f(x) zadana jako funkcja anonimowa
        x0, x1: przybliżenia początkowe
        delta, epsilon: dokładność obliczeń 
        maxit: maksymalna liczba iteracji
    
    Dane wyjściowe w postaci: (r,v,it,err)
        r: szukane przybliżenie
        v: wartość funkcji w punkcie r
        it: liczba iteracji
        err: informacja o błędzie:
            0: brak błędu, metoda zbieżna
            1: nie osiągnięto wymaganej dokładności w podanej liczbie iteracji 
    
    """ 
    
    f_x0 = f(x0)
    f_x1 = f(x1)
    
    for it = 1:maxit
        
        if abs(f_x0) > abs(f_x1)
            x0, x1 = x1, x0
            f_x0, f_x1 = f_x1, f_x0
        end
        
        s = (x1 - x0)/(f_x1 - f_x0)
        x1 = x0
        f_x1 = f_x0
        x0 = x0 - f_x0 * s
        f_x0 = f(x0)
        
        if abs(x1 - x0) < delta || abs(f_x0) < epsilon 
            return (x0, f_x0, it, 0)
        end
    end
    
    return (x0, f_x0, maxit, 1)
    
end
```

### Opis

Metoda siecznych, podobnie jak metoda Newtona, jest metodą iteracyjną. Również korzysta z pochodnej funkcji $f$, jednak tym razem nie trzeba podawać jej dokładnej wartości. Pochodną aproksymujemy za pomocą następującego wzoru: 
$f'(x_n) \approx \frac{f(x_n) - f(x_{n-1})}{x_n - x_{n-1}}$. Jeśli chodzi o przybliżenia pierwiastków, stosowany jest wzór, wynikający bezpośrednio z poprzedniego: $x_{n+1} = x_{n} - \frac{x_n - x_{n-1}}{f(x_n) - f(x_{n-1})}f(x_n)$ dla $n \ge 1$. W tym przypadku będziemy potrzebowali dwóch przybliżeń początkowych na wejściu - $x_0$ i $x_1$. 

Działanie algorytmu jest bardzo podobne do metody Newtona, z tą różnicą, że pochodna jest wyliczana w trakcie iteracji. Swoje działanie rozpoczyna licząc wartości funkcji w podanych punktach (przybliżeniach początkowych). Następnie wykonywana jest pętla, również $maxit$ razy. Na początku sprawdzany jest warunek: $|f_{x_0}|$ > $|f_{x_1}|$. Jeżeli jest prawdziwy, to następuje "swap" wartości zmiennych (niezbędne, aby uzyskać warunek końca). W przeciwnym wypadku obliczany jest pierwiastek funkcji zgodnie z podanym wcześniej wzorem (rozbitym na 2 linijki) i dokonywana odpowiednia zamiania wartości zmiennych. Następnie sprawdzany jest warunek końca: odległość między punktami $x_0$ i $x_1$ jest mniejsza niż delta lub wartość funkcji w nowo wyliczonym punkcie jest mniejsza niż zadana precyzja. W przypadku gdy żaden z warunków nie jest spełniony wykonywana jest kolejna iteracja. Jeżeli  maksymalna liczba iteracji zostanie przekroczona, zwracane są: ostatnie przybliżenie pierwiastka oraz kod błędu: 1

##  Zadanie 6

Znaleźć miejsce zerowe funkcji $f_1(x) = e^{1−x}− 1$ oraz $f_2(x) =xe^{−x}$ za pomocą metod bisekcji, Newtona i siecznych. 

```julia
using Plots
f1(x) = ℯ^(1-x) - 1 
f2(x) = x*ℯ^(-x) 
f0(x) = 0
args = [Float64(i/100) for i = -1200:1200]
plot(args, [f1, f0], label=["f1(x)" "x = 0"], lw=2)
```

```julia
plot(args,[f2,f0], label=["f2(x)" "x = 0"], lw=2)
```

Na podstawie powyższych wykresów można zauważyć, że miejsce zerowe funckji $f_1(x) = e^{1−x}− 1$ znajduje się w x = 1, a $f_2(x) =xe^{−x}$ dla x = 0. Sprawdźmy czy zaimplementowane algorytmy zwracają poprawne wyniki.

```julia
starts = [(0.0, 2.0), (-0.5, 1.5), (-100.0, 100.0), (0.1, 1.8)]


println("Metoda bisekcji dla f1(x)")
println("Przedział; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in starts
    mb = mbisekcji(f1, el[1], el[2], Float64(10^-5), Float64(10^-5)) 
    println(el, " ", mb)
end

starts = [-1.0, 0.0, 0.9, -10.0, 4.0, 7.0, 100.0,  8.0]
println("\nMetoda stycznych (Newtona) dla f1(x)")
println("Przybliżenie początkowe; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in starts
    ms = mstycznych(f1, x -> -ℯ^(1.0-x), el, Float64(10^-5), Float64(10^-5), 1000)
    println(el, " ", ms)
end

starts = [(-2.0, 2.0), (-1.0, 1.5), (0.1, 1.3), (0.0, 12.0)]
println("\nMetoda siecznych dla f1(x)")
println("Przybliżenia początkowe; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in starts
    msi = msiecznych(f1, el[1], el[2], Float64(10^-5), Float64(10^-5), 40)
    println(el, " ", msi)
end
```

Każdy algorytm wyznaczył miejsce zerowe z wymaganą dokładnością, dla odpowiednich parametrów wejściowych

Metoda bisekcji, zawsze zwraca poprawny wynik, a w przypdaku dobrze dobranego przedziału (miejsce zerowe znajduje się w jego połowie) jest to wynik dokładny (0.0). 

W przypadku metody Newtona, wyznaczenie miejsca zerowego jest znacznie szybsze niż w metodzie bisekcji (wyjątkiem są dobrze dobrane przedziały, o których wspomniałem wyżej). Nie jest to jednak metoda idealna; przekazanie przybliżenia początkowego będącego w przedziale $(1, \infty)$ powoduje znaczny wzrost potrzebnych iteracji, a od pewnego momentu błąd - NaN (Not a number).

W metodzie stycznych dużą rolę odgrywa wybranie odpowiednich przybliżeń początkowych. Dla bliskiego otoczenia miejsca zerowego, wyniki są prawidłowe. W przypadku wartości odległych od miejsca zerowego, algorytm nie jest w stanie wyznaczyć wartości z podaną dokładnością, przez co wynik jest daleki od rzeczywistego.

```julia
intervals = [(-1.0, 0.5), (-0.5, 0.5), (-100.0, 100.0)]

println("Metoda bisekcji dla f2(x)")
println("Przedział; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in intervals
    mb = mbisekcji(f2, el[1], el[2], Float64(10^-5), Float64(10^-5)) 
    println(el, " ", mb)
end

starts = [-1.0, 0.0, 0.9, -10.0, 1.5, 4.0, 10.0, 1.0]

println("\nMetoda stycznych (Newtona) dla f2(x)")
println("Przybliżenie początkowe; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in starts
    ms = mstycznych(f2, x -> -ℯ^(-x) * (x - 1.0), el, Float64(10^-5), Float64(10^-5), 100)
    println(el, " ", ms)
end

starts = [(0.1, 1.3), (-0.5, 0.5), (-5.0, 0.5), (-2.0, 2.0), (-100.0, 100.0)]
println("\nMetoda siecznych dla f2(x)")
println("Przybliżenia początkowe; x0; f(x0); liczba iteracji; błąd")
println("---------------------------------------------------------")
for el in starts
    msi = msiecznych(f2, el[1], el[2], Float64(10^-5), Float64(10^-5), 40)
    println(el, " ", msi)
end
```

Efekt wyznaczania miejsca zerowego $f_2(x)$ jest podobny jak dla poprzedniej funkcji. 

Metoda bisekcji zwróciła poprawne wyniki, również przedział nie miał znaczenia (dobry przedział początkowy gwarantował dokładną wartość).

Metoda Newtona zwróciła ciekawe wyniki dla przybliżeń początkowych > 1; x ~ 14 jest dalekie od faktycznego miejsca zerowego, jednak prawidłowe pod względem zadanego błędu. Z kolei dla x = 1, pochodna rozważanej funkcji jest bliska 0, więc algorytm od razu kończy działanie z kodem błędu = 2.

Metoda siecznych ponownie miała problem z dalekim otoczeniem $x_0$, jednak wartości funkcji w otrzymanych wynikach tj. $x = 100$ i $x = 14.294924723787231$ również mieszczą się w zadanym błędzie. 

W przypadku metod Newtona oraz siecznych, wybór parametrów początkowych okazał się kluczowy. 

Na podstawie powyższych eksperymentów potwierdza się fakt, że metoda bisekcji jest najstabilniejsza (zbieżność globalna), ale i najwolniejsza. Dwie pozostałe metody są o wiele szybsze (zbieżność lokalna), ale w przypadku źle dobranych parametrów początkowych mogą zwrócić niezadowalające nas wyniki.
